```python
import matplotlib.pyplot as plt
```

```python
# Scatter Plot

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

plt.title("Scatter plot", fontsize=12)
plt.xlabel("Day", fontsize=10)
plt.ylabel("Number", fontsize=10)

plt.scatter(day, no, color="r")

plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

plt.title("Scatter plot", fontsize=12)
plt.xlabel("Day", fontsize=10)
plt.ylabel("Number", fontsize=10)

c = ["r", "y", "g", "b", "r", "g", "m" ]
sizes = [400, 200, 100, 300, 500, 100, 600]
plt.scatter(day, no, color=c, s=sizes, alpha=0.6, marker="*", edgecolor="r", linewidth=2)

plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

plt.title("Scatter plot", fontsize=12)
plt.xlabel("Day", fontsize=10)
plt.ylabel("Number", fontsize=10)

c = [10, 49, 30, 29, 56, 20, 30]
sizes = [400, 200, 100, 300, 500, 100, 600]
plt.scatter(day, no, c=c, s=sizes, cmap="viridis")
# colormap -> cmap -> it is only run when you c=c kroge -> rest can see documantaion
t = plt.colorbar()
t.set_label("Color Bar")

plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
no2 = [3, 2, 4, 5, 1, 6, 2]

plt.title("Scatter plot", fontsize=12)
plt.xlabel("Day", fontsize=10)
plt.ylabel("Number", fontsize=10)

c = [10, 49, 30, 29, 56, 20, 30]
sizes = [400, 200, 100, 300, 500, 100, 600]
plt.scatter(day, no, c=c, s=sizes, cmap="viridis")
plt.scatter(day, no2, c="r", s=sizes)
t = plt.colorbar()
t.set_label("Color Bar")

plt.show()
```
