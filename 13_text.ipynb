```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x,y, color="b")

plt.title("Plot Graph", fontsize = 15)
plt.xlabel("Days", fontsize = 12)
plt.ylabel("Python", fontsize = 12)

plt.text(3,5, "Avg. reach",color="w", fontsize = 12, style="italic", bbox={"facecolor":"g", "alpha":0.6})

plt.annotate("High peak Python", xy=(5,6), xytext=(3,6), arrowprops=dict(facecolor="black", shrink=200))

plt.legend(["up"], loc=2, facecolor="w", edgecolor="b", framealpha=0.5, shadow=False)

plt.show()
```
