```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60, 82]

plt.xlabel("Programming language", fontsize=12)
plt.ylabel("People votes", fontsize=12)
plt.title("Popularity of Language", fontsize=15)

c = ["y", "b", "m", "g"]
plt.bar(x, y, 
        width=0.5, color=c, align="center", 
        edgecolor='#000', linewidth=3, linestyle="--", alpha=0.6)
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60, 82]
z = [20, 30, 40, 50]
plt.xlabel("Programming language", fontsize=12)
plt.ylabel("People votes", fontsize=12)
plt.title("Popularity of Language", fontsize=15)
plt.bar(x, y, 
        width=0.5, color='r',  
        label = "Popularity")
plt.bar(x, z, 
        width=0.5, color='y',  
        label = "Popularity1")
plt.legend()
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60, 82]
z = [20, 30, 40, 50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Programming language", fontsize=12)
plt.ylabel("People votes", fontsize=12)
plt.title("Popularity of Language", fontsize=15)

plt.bar(p, y, 
        width, color='r',  
        label = "Popularity")
plt.bar(p1, z, 
        width, color='y',  
        label = "Popularity1")

plt.xticks(p+width/2, x, rotation=20)
plt.legend()
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60, 82]
z = [20, 30, 40, 50]

plt.xlabel("Programming language", fontsize=12)
plt.ylabel("People votes", fontsize=12)
plt.title("Popularity of Language", fontsize=15)

plt.barh(x, y, 
        color='r',  
        label = "Popularity")
plt.barh(x, z, 
        color='y',  
        label = "Popularity1")

# for horizontal graph just write h in plt.barh

plt.legend()
plt.show()
```
