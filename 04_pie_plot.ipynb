```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40]
y = ["c", "c++", "java", "python"]
ex = [0.3, 0, 0, 0]
c = ["r", "b", "g", "y"]

plt.pie(x,labels=y, explode=ex, autopct="%0.1f%%", shadow=True, radius=1, labeldistance=1.1, startangle=190, 
        textprops={"fontsize": 15}, counterclock=False, wedgeprops={"linewidth": 2, "edgecolor": "m"}, rotatelabels=False)

plt.title("Language")
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
# Dot pie chart
```

```python
x = [10, 20, 30, 40]
x1 = [40, 30, 20, 10]
y = ["c", "c++", "java", "python"]



plt.pie(x,labels=y, radius=1.5)
plt.pie(x1, radius=0.7)


plt.show()
```

```python
x = [10, 20, 30, 40]
x1 = [40, 30, 20, 10]
y = ["c", "c++", "java", "python"]



plt.pie(x,labels=y, radius=1.5)
plt.pie([1], colors="w")


plt.show()
```

```python
x = [10, 20, 30, 40]
x1 = [40, 30, 20, 10]
y = ["c", "c++", "java", "python"]



plt.pie(x,labels=y, radius=1.5)
cr = plt.Circle(xy=(0,0), radius=1, facecolor="w")
plt.gca().add_artist(cr)


plt.show()
```
