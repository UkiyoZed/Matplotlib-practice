```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40]
y = ["c", "c++", "java", "python"]
ex = [0.3, 0, 0, 0]
c = ["r", "b", "g", "y"]

plt.pie(x,labels=y, explode=ex, autopct="%0.1f%%", shadow=True, radius=1.3, labeldistance=1.1, startangle=190, 
        textprops={"fontsize": 15}, counterclock=False, wedgeprops={"linewidth": 2, "edgecolor": "m"})

plt.show()
```
