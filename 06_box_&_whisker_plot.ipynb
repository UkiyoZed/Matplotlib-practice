```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, notch=False, vert=True, widths=0.1, labels=["Python"], patch_artist=False, 
            showmeans=True, whis=1, sym="r+", boxprops=dict(color="r"), capprops=dict(color="b"), whiskerprops=dict(color="g"),
            flierprops=dict(markeredgecolor="y"))

plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]
x1 = [10, 20, 40, 50, 30, 60, 90]


plt.boxplot(x, notch=False, vert=True, widths=0.1, labels=["Python"], patch_artist=False, 
            showmeans=True, whis=1, sym="r+", boxprops=dict(color="r"), capprops=dict(color="b"), whiskerprops=dict(color="g"))

plt.boxplot(x1, notch=False, vert=True, widths=0.1, labels=["Python"], patch_artist=False, 
            showmeans=True, whis=1, sym="r+", boxprops=dict(color="r"), capprops=dict(color="b"), whiskerprops=dict(color="g"))

plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]
x1 = [10, 20, 40, 50, 30, 60, 90]

y = [x, x1]

plt.boxplot(y,labels=["Python", "c++"],  
            showmeans=True, sym="g+", boxprops=dict(color="r"), capprops=dict(color="b"), whiskerprops=dict(color="g"))



plt.show()
```
