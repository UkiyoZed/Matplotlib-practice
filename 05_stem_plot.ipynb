```python
import matplotlib.pyplot as plt
```

Character -> '-' (solid line), '--' (Dashed line), '-.' (Dash-dot line), ':' (Dotted line)

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x,y, linefmt="-.", markerfmt="r+", bottom=0, basefmt="g", label="Python", orientation="horizontal")

plt.legend()
plt.show()
```
